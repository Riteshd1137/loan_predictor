```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")
```

```python
df = df[["ApplicantIncome", "Education", "Credit_History", "Loan_Status"]]
df.dropna(inplace=True)
```

```python
df["Education"] = df["Education"].map({"Graduate": 1, "Not Graduate": 0})
df["Loan_Status"] = df["Loan_Status"].map({"Y": "Eligible", "N": "Not Eligible"})
```

```python
X = df[["ApplicantIncome", "Education", "Credit_History"]]
y = df["Loan_Status"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("âœ… Model Accuracy:", accuracy)
```

```python
plt.figure(figsize=(18, 12))
plot_tree(
    clf,
    filled=True,
    feature_names=["ApplicantIncome", "Education", "Credit_History"],
    class_names=["Not Eligible", "Eligible"],
    rounded=True,
    fontsize=10
)
plt.title("Loan Eligibility â€“ Decision Tree")
```
